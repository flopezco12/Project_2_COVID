```python
vaccinations_metadatadf = pd.read_csv('../CSV_files/vaccinationmetadata.csv')
vaccinations_metadatadf.head(10)
```

```python
columns_to_keep = [ 'VACCINE_NAME', 'PRODUCT_NAME', 'COMPANY_NAME']

vaccines_metadata_cleaned_df = vaccinations_metadatadf[columns_to_keep]

vaccines_metadata_cleaned_df
```

```python
vaccinations_metadatadf.drop(vaccinations_metadatadf.columns.difference(columns_to_keep), axis=1, inplace=True)

vaccinations_metadatadf

```

```python
vaccinations_metadatatest = vaccinations_metadatadf

vaccinations_metadatatest
```

```python
vaccinations_meta_dropped = vaccinations_metadatatest.dropna() 
vaccinations_meta_dropped
```

```python
vacc_df = vaccinations_meta_dropped.groupby('PRODUCT_NAME')['COMPANY_NAME'].unique()
vaccines_df = vacc_df.apply(pd.Series)
vaccines_df 

vaccines_reset_df = vaccines_df.reset_index()
vaccines_reset_df
```
