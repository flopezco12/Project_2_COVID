```python
vaccinations_metadatadf = pd.read_csv('../CSV_files/vaccinationmetadata.csv')
vaccinations_metadatadf.head(10)
```

```python
columns_to_keep = [ 'VACCINE_NAME', 'PRODUCT_NAME', 'COMPANY_NAME']

vaccines_metadata_cleaned_df = vaccinations_metadatadf[columns_to_keep]

vaccines_metadata_cleaned_df
```

```python
vaccinations_metadatadf.drop(vaccinations_metadatadf.columns.difference(columns_to_keep), axis=1, inplace=True)

vaccinations_metadatadf

```

```python
vaccinations_metadatatest = vaccinations_metadatadf

vaccinations_metadatatest
```

```python
vaccinations_meta_dropped = vaccinations_metadatatest.dropna() 
vaccinations_meta_dropped
```

```python
vaccinations_meta_dropped.nunique()
```

```python
vacc_df = vaccinations_meta_dropped.groupby('PRODUCT_NAME')['COMPANY_NAME'].unique()
vaccines_df = vacc_df.apply(pd.Series)
vaccines_df 

vaccines_reset_df = vaccines_df.reset_index()
vaccines_reset_df
```

```python
vaccines_final_df = vaccines_reset_df.rename(columns = {vaccines_reset_df.columns[1]: 'Company_name'})
vaccines_final_df
```

```python
#FINAL VACCINE META DF 
vaccines_final_df
```

```python
vaccines_final_df.to_csv("../CSV_files/FINAL1cleanedvaccinationmetadata.csv", encoding='utf8', index=False)
```

```python
import psycopg2
#open connection to database
conn_string = """host=localhost 
                dbname=Project_COVID_db 
                user=postgres 
                password=postgres"""
conn = psycopg2.connect(conn_string)
cursor = conn.cursor()
print("Connected!")
```

```python
#Drop tables if they exist
cursor.execute("DROP TABLE IF EXISTS Dailycases")
cursor.execute("DROP TABLE IF EXISTS Vaccinations")
cursor.execute("DROP TABLE IF EXISTS VaccineMetaData")
```

```python
#Create tables
cursor.execute("""
    CREATE TABLE "Dailycases" (
        Date_reported date PRIMARY KEY,
        Country VARCHAR,
        New_cases INTEGER,
        Cumulative_cases INTEGER,
        New_deaths Integer,
        Cumulative_deaths INTEGER
    )
""")

cursor.execute("""
    CREATE TABLE  "Vaccinations" (
        COUNTRY VARCHAR PRIMARY KEY,
        DATE_UPDATED date,
        TOTAL_VACCINATIONS INTEGER,
        PERSONS_VACCINATED_1PLUS_DOSE INTEGER,
        TOTAL_VACCINATIONS_PER100 decimal,
        PERSONS_VACCINATED_1PLUS_DOSE_PER100 decimal,
        PERSONS_LAST_DOSE INTEGER,
        PERSONS_LAST_DOSE_PER100 decimal,
        FIRST_VACCINE_DATE date,
        NUMBER_VACCINES_TYPES_USED INTEGER,
        PERSONS_BOOSTER_ADD_DOSE INTEGER,
        PERSONS_BOOSTER_ADD_DOSE_PER100 decimal
    )
""")

cursor.execute("""
    CREATE TABLE "VaccineMetaData" (
        PRODUCT_NAME VARCHAR PRIMARY KEY,
        Company_name VARCHAR
        
    )
""")
```
